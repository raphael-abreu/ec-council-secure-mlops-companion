```python
gcloud projects create secure-mlops-$(date +%s) --name="Secure MLOps Course"
PROJECT_ID=$(gcloud config get-value project)

# 2. Enable APIs
gcloud services enable \
  aiplatform.googleapis.com \
  iam.googleapis.com \
  storage.googleapis.com \
  containerregistry.googleapis.com

# 3. Create secure bucket
BUCKET_NAME="secure-mlops-$(date +%s)"
gsutil mb -p $PROJECT_ID gs://$BUCKET_NAME

echo "‚úÖ GCP environment ready: $PROJECT_ID"
echo "üì¶ Secure bucket: gs://$BUCKET_NAME"
```

```python
gcloud iam service-accounts create secure-ml-pipeline \
  --display-name="Secure ML Pipeline Service Account"

SA_EMAIL="secure-ml-pipeline@$PROJECT_ID.iam.gserviceaccount.com"

gcloud projects add-iam-policy-binding $PROJECT_ID \
  --member="serviceAccount:$SA_EMAIL" \
  --role="roles/aiplatform.user"

gcloud projects add-iam-policy-binding $PROJECT_ID \
  --member="serviceAccount:$SA_EMAIL" \
  --role="roles/storage.objectAdmin"

echo "üîê Service account created with least-privilege access"
```
